```python
import numpy as np
np.random.seed(0)

def str_to_list(cadena):
    """
    Parameters
    --------------

    cadena : string con el texto a analizar solo con las mayusculas

    Brief
    --------------
    Funcion para transformar las cadenas de texto en arrays numéricos
    señalando el número al que pertenecen.
    """

    l = []
    
    for s in cadena:
        x = ord(s)
        if x < 91:
            l.append(x-65)

    return l


def list_to_str(lista):
    
    """
    Parameters
    --------------

    lista : lista de enteros entre 0 y 25

    Brief
    --------------
    Funcion para transformar los arrays numéricos que señalan el número 
    al que pertenecen los caracteres a una cadena. 
    """
    
    s=""
    
    for i in lista:
        if -1< int(i) and int (i) <26:
            x = chr(int(i)+65)
            s += ' ' + x 

    return s

def text_to_matrix(txt, n_rows):
    """
    Parameters
    -------------
    txt : Texto de entrada
    n_rows: Numero de filas de la matriz de cifrado
    
    Brief
    -------------
    Funcion que transforma el mensaje de texto a una matriz numerica 
    haciendo uso de la funcion str_to_list

"""
    l=str_to_list(txt)
    
    n_columns = ceil(len(l)/n_rows) 
    while len(l)!=n_rows*n_columns:
        # Si el texto no lo podemos poner en forma matricial añadimos
        # caracteres aleatorios entre 0 y 26
        l.append(l[0])
    # Reescalamos ahora la matriz a tamaño n_rows, n_columns.
    # Para ello primero reescalamos en n_columns,n_rows y luego
    # calculamos la traspuesta
    return np.array(l).reshape((n_rows,n_columns), order='F')

def matrix_to_text(matrix):
    """
        Parameters
        -------------
        matrix : Matriz de entrada

        Brief
        -------------
        Funcion que transforma la matriz numerica en texto haciendo 
        uso de la funcion list_to_str
    """
    # Cambio la matriz a un vector (cambio por columnas)
    l = matrix.flatten(order='F')
    return list_to_str(l)
```
